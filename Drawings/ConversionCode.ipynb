```python
import pandas as pd
import numpy as np
import os
print(pd.__version__)
```

```python
a = np.zeros((18, 18), dtype=np.int32)
files = os.listdir()
print(files)
```

```python
for file in files:
    if not ("ipynb" in file or "csv" in file):
        try:
            df = pd.read_csv(file)
            a = np.zeros((17, 17), dtype=int)
            for x in range(0, 17):
                for y in range(0, 17):
                    if int(df.iloc[2 + x * 18 + y]["P2"]) < 128:
                        a[x][y] = 1
            output_filename = os.path.splitext(file)[0] + ".csv"
            np.savetxt(output_filename, a, delimiter=",", fmt="%d")
        except Exception as e:
            print(f"{file}: {e}")
```
