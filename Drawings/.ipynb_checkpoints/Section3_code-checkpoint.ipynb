```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
df = pd.read_csv("40403863_features.csv")
for x in range(2,8):
    df.iloc[:, x].plot(kind="hist", edgecolor="black")
    plt.xlabel(df.iloc[:,x].name)
    plt.savefig(df.iloc[:,x].name + "graph", dpi=300, bbox_inches='tight')
    plt.show()
    plt.clf()
```
