```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
df = pd.read_csv("40403863_features.csv")
#struggled to figure out how to split into letters and non letters, found the filter method in the pandas documentation
for x in range (2, 18):
    column = df.columns[x]
    letters = df[column].head(80)
    non_letters = df[column].tail(60)
    
    print(column + "\n")
    print("Letters:\t", letters.mean())
    print("Non Letters:\t", non_letters.mean())
    print(df[column].median())
    print(df[column].std())
    print("\n")
```
