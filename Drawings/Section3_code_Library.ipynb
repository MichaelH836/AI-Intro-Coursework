```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
df = pd.read_csv("40403863_features.csv")
#struggled to figure out how to split into letters and non letters, found the filter method in the pandas documentation
for x in range (2, 18):
    column = df.columns[x]
    letters = df[column].head(80)
    non_letters = df[column].tail(60)
    print(column)
    print("MEAN -", "Letters: ", letters.mean().round(2), "Non-Letters: ", non_letters.mean().round(2))
    print("MEDIAN -", "Letters: ", letters.median().round(2), "Non-Letters: ", non_letters.median().round(2))    
    print("SD -", "Letters: ", letters.std().round(2), "Non-Letters: ", non_letters.std().round(2))
    print("\n")
```
